**Exercise-3 Print a pie chart of revenue realized per booking platform**

```python
df_bookings_all.groupby("booking_platform")["revenue_realized"].sum().plot(kind="pie")
```

```python
import matplotlib.pyplot as plt

# Grouped data
revenue_sum = df_bookings_all.groupby("booking_platform")["revenue_realized"].sum()

# Pie chart with values + percentage
revenue_sum.plot(
    kind="pie",
    autopct=lambda p: f'{p:.1f}%\n({p*revenue_sum.sum()/100:.0f})',  # % + value
    figsize=(7,7),
    ylabel=""
)

plt.title("Revenue by Booking Platform")
plt.show()

# 1. lambda p: ...
# Ye ek anonymous function (lambda function) hai.
# Jab bhi pie chart ke ek slice ki percentage calculate hoti hai, wo percentage value p ke andar aati hai.
# Example: agar ek slice 25% hai to p = 25.0 hoga.
# 2. f'{p:.1f}%'
# Ye ek f-string formatting hai.
# {p:.1f} ka matlab hai ki p ko 1 decimal place tak round karke likho.
# Agar p = 25.345 hoga to output "25.3".
# '% lagane se text ban jayega "25.3%".
# 3. \n
# Ye newline character hai.
# Matlab pie chart ke label mein do line banengi:
# Pehli line: percentage (e.g. 25.3%)
# Dusri line: actual revenue value (neeche samajhte hain).
# 4. (p*revenue_sum.sum()/100:.0f)
# Yaha percentage ko actual number me convert kiya ja raha hai.
# revenue_sum.sum() = total revenue ka sum (maan lo total revenue = 10,000).
# Agar slice 25% hai (p = 25):
# p * revenue_sum.sum() / 100
# = 25 * 10000 / 100
# = 2500
# {:.0f} ka matlab hai result ko decimal ke bina integer banake show karo.
# "2500"
# 5. Final Output
# Agar ek slice 25% hai aur total revenue 10,000 hai, to label banega:
# 25.0%
# (2500)
# Pehli line percentage hogi, dusri line uska actual revenue value.
```
